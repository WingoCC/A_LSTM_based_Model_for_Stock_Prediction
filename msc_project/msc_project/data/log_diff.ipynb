```python
price = data['AMZ_Close']
price
```

```python
price_log = np.log(price)
```

```python
price_log
```

```python
new_data = pd.DataFrame()
```

```python
new_data['log'] = price_log.tolist()
```

```python
price_log = price_log
diff = []
for i in range(len(price_log)-1):
    b = price_log[i+1] - price_log[i]
    diff.append(b)
```

```python
diff
```

```python
diff.insert(0,0)
```

```python
data['log_diff'] = diff
```
