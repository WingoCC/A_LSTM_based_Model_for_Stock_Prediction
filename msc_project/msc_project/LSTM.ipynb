```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from collections import deque
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Dropout
```

```python
#处理数据
def Process_data(data):
    # 数据标准化
    scaler = StandardScaler()
    sca_X = scaler.fit_transform(data.iloc[:,:-1])

    men_his_day = 21 #设置记忆天数 先尝试21即一个月的交易日

    #队列
    dep = deque(maxlen=men_his_day)

    X = []

    for i in sca_X:
        dep.append(list(i))
        if len(dep) == men_his_day:
            X.append(list(dep))


    Y = data['AMZ_Close'].values

    X = np.array(X)
    Y = np.array(Y)
    
    X_train,X_test,Y_train,Y_test = X[:2012],X[2012:],Y[:2012],Y[2012:]
    
    return X_train,X_test,Y_train,Y_test

X_train,X_test,Y_train,Y_test = Process_data(date)
```

```python
def pre(X_train,X_test,Y_train,Y_test):
    model = Sequential()

    #构建3层神经网络

    #设置神经单元
    model.add(LSTM(10,input_shape= X_train.shape[1:],activation='relu',return_sequences=True))
    #防止过拟合，删除10%的神经元
    model.add(Dropout(0.1))

    # 设置神经单元
    model.add(LSTM(32,activation='relu', return_sequences=True))
    # 防止过拟合，删除10%的神经元
    model.add(Dropout(0.1))

    # 设置神经单元
    model.add(LSTM(32,activation='relu'))
    # 防止过拟合，删除10%的神经元

    model.add(Dropout(0.1))
    #构建全连接层
    model.add(Dense(32,activation='relu'))
    model.add(Dropout(0.1))

    #输出层
    model.add(Dense(1))

    model.compile(optimizer='adam',loss='mse',metrics=['mape'])
    #参考网址:https://blog.csdn.net/weixin_39559994/article/details/105842315

    #训练模型
    model.fit(X_train,Y_train,batch_size=32,epochs=50,validation_data=(X_test,Y_test))
    #参考网址:https://www.cnblogs.com/wwyt123/p/14896713.html
    pre = model.predict(X_test)
    return pre ,Y_test
```

```python
pre ,Y_test = pre(X_train,X_test,Y_train,Y_test)
df_time = date.index[-len(X_test):]
plt.plot(df_time,Y_test,label='true')
plt.plot(df_time,pre,label='predict')
plt.legend()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
lstm_layers = [1,2,3]
dense_layers = [1,2,3]
units = [16,32]

for the_lstm_layers in lstm_layers:
    for the_dense_layers in dense_layers:
        for the_units in units:
            #'C:\Users\lenovo_\Desktop'
            filepath = r'C:\Users\lenovo_\Desktop\model\{val_mape:.2f}_{epoch:02d}_'+ \
            f'lstm_{the_lstm_layers}_dense_{the_dense_layers}_unit_{the_units}.h5'
            checkpoint = ModelCheckpoint(
                filepath,
                save_weights_only = False,
                monitor = 'val_mape',
                mode = 'min',
                save_best_only = False
                )
            
            model = Sequential()

            #构建3层神经网络

            #设置神经单元
            model.add(LSTM(10,input_shape= X_train.shape[1:],activation='relu',return_sequences=True))
            #防止过拟合，删除10%的神经元
            model.add(Dropout(0.1))
            
            for i in range(the_lstm_layers):
                # 设置神经单元
                model.add(LSTM(the_units,activation='relu', return_sequences=True))
                # 防止过拟合，删除10%的神经元
                model.add(Dropout(0.1))

            # 设置神经单元
            model.add(LSTM(the_units,activation='relu'))
            # 防止过拟合，删除10%的神经元
            
            model.add(Dropout(0.1))
            for i in range(the_dense_layers):
                #构建全连接层
                model.add(Dense(the_units,activation='relu'))
                model.add(Dropout(0.1))

            #输出层
            model.add(Dense(1))

            model.compile(optimizer='adam',loss='mse',metrics=['mape'])
            #参考网址:https://blog.csdn.net/weixin_39559994/article/details/105842315

            #训练模型
            model.fit(X_train,Y_train,batch_size=32,epochs=50,validation_data=(X_test,Y_test),callbacks=[checkpoint])
            #参考网址:https://www.cnblogs.com/wwyt123/p/14896713.html
```
