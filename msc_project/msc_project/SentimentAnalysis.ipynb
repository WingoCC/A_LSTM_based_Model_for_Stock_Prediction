```python
def data_preprocess(df, column_name='content'):
    df_ret = deepcopy(df)
    
    df_ret = df_ret.dropna()
    
    df_ret[column_name] = df_ret[column_name].apply(lambda x: re.sub('<[^<]+?>', ' ', x))
    df_ret[column_name] = df_ret[column_name].str.replace('[^\w\s]', ' ', regex=True)
    
    df_ret[column_name] = df_ret[column_name].apply(lambda x: contractions.fix(x))
    
    df_ret[column_name] = df_ret[column_name].apply(lambda x: " ".join(item.lower() for item in x.split()))
    
#     split the sentences into tokens
    df_ret[column_name] = df_ret[column_name].apply(lambda x: nltk.word_tokenize(x))
    
#     remove stopwords
    df_ret[column_name] = df_ret[column_name].apply(lambda x: " ".join(item for item in x if item not in stopwords.words("english")))
    
#     convert tokens to their root form
    lemmatizer = WordNetLemmatizer()
    df_ret[column_name] = df_ret[column_name].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()]))
    
#     reset row index
    df_ret = df_ret.reset_index(drop=True)
    
    return df_ret
    
```

```python
clean_news = data_preprocess(news)
```
